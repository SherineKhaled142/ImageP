# Mean Filter

```python
def mean_filter(image, kernel_size):
    padded_image = cv2.copyMakeBorder(image,
                                      kernel_size//2, 
                                      kernel_size//2, 
                                      kernel_size//2,
                                      kernel_size//2,
                                      cv2.BORDER_CONSTANT)
    
    filtered_image = np.zeros_like(image)
    
    for y in range(image.shape[0]):
        for x in range(image.shape[1]):
            window = padded_image[y:y+kernel_size, x:x+kernel_size]
            mean_value = np.mean(window)
            filtered_image[y, x] = mean_value
    
    return filtered_image
```

```python
mean_filtered = mean_filter(img,5)

plt.title('Mean filter')
plt.imshow(mean_filtered, cmap='gray')
plt.axis('off')
```

# Median Filter

```python
def median_filter(image, kernel_size):
    padded_image = cv2.copyMakeBorder(image, kernel_size // 2, kernel_size // 2, kernel_size // 2, kernel_size // 2, cv2.BORDER_REFLECT)

    filtered_image = np.zeros_like(image)

    for y in range(image.shape[0]):
        for x in range(image.shape[1]):
            window = padded_image[y:y + kernel_size, x:x + kernel_size]
            median_value = np.median(window)
            filtered_image[y, x] = median_value

    return filtered_image
```

```python
median_filtered = median_filter(img,3)

plt.title('Median filter')
plt.imshow(median_filtered, cmap='gray')
plt.axis('off')
```
